```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,LSTM
```

```python
#load dataset
dataset=pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/AirPassengers.csv")
```

```python
data=dataset["value"].values
data=data.reshape(-1,1)
```

```python
#normalise the data
scaler=MinMaxScaler(feature_range=(0,1))
data=scaler.fit_transform(data)
```

```python
#create time series data
time_Step=5
```

```python
x=[]
y=[]
```

```python
for i in range(len(data)-time_Step-1):
   x.append(data[i:i+time_Step])
   y.append(data[i+time_Step,0])
```

```python
x=np.array(x)
y=np.array(y)
```

```python
#reshape x values
x=x.reshape(x.shape[0],x.shape[1],1)
```

```python
#RNN model
rnn_model=Sequential()
rnn_model.add(SimpleRNN(50,activation='tanh',input_shape=(x.shape[1],1)))
rnn_model.add(Dense(1))
rnn_model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
#train model
rnn_model.fit(x,y,epochs=50,batch_size=16)
```

```python
#evaluate RNN model
rnn_pred=rnn_model.evaluate(x,y)
```

```python
#LSTM model
lstm_model=Sequential()
lstm_model.add(LSTM(50,input_shape=(time_Step,1)))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
#train lstm model
lstm_model.fit(x,y,epochs=50,batch_size=16)
```

```python
#evaluate model
lstm_pred = lstm_model.evaluate(x, y, verbose=0)
print(f"LSTM Model Loss (MSE): {lstm_pred}")

# You can also make predictions and inverse transform them to see actual values
lstm_train_predict = lstm_model.predict(x)
lstm_train_predict = scaler.inverse_transform(lstm_train_predict)

# Assuming y_actual was already computed for RNN evaluation or can be re-computed
y_actual = scaler.inverse_transform(y.reshape(-1, 1))

import math
from sklearn.metrics import mean_squared_error
lstm_rmse = math.sqrt(mean_squared_error(y_actual, lstm_train_predict))
print(f"LSTM Train RMSE: {lstm_rmse}")
```
