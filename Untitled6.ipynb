Early Stopping

```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.layers import Dense,Input,Flatten
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#load datasets
(x_train,y_train),(x_test,y_test)=fashion_mnist.load_data()
```

```python
#normalize
x_train=x_train/255
x_test=x_test/255
```

```python
#reducing the trained data size(to create overfitting(small dataset over large network))
x_train_small,_,y_train_small,_=train_test_split(
   x_train,y_train,train_size=6000,random_state=42
)
```

```python
#holdout method for splitting trained data for training and validation
x_train,x_val,y_train,y_val=train_test_split(
    x_train_small,y_train_small,train_size=0.8,random_state=42
)
```

```python
#define model
inputs=Input(shape=(28,28))
x=Flatten()(inputs)
x=Dense(1024,activation='relu')(x)
x=Dense(512,activation='relu')(x)
x=Dense(256,activation='relu')(x)
outputs=Dense(10,activation='softmax')(x)
#model wiothput early stopping
model1=Model(inputs,outputs)
```

```python
#compile the model
model1.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#train the model
history1=model1.fit(
    x_train,y_train,epochs=40,batch_size=64,validation_data=(x_val,y_val),verbose=1
)
```

```python
#evaluate the model
model1.loss,model1_accuracy=model1.evaluate(x_test,y_test)
print('Accuracy',model1_accuracy)
```

```python
#model with early stopping
inputs1=Input(shape=(28,28))
x1=Flatten()(inputs1)
x1=Dense(1024,activation='relu')(x1)
x1=Dense(512,activation='relu')(x1)
x1=Dense(256,activation='relu')(x1)
outputs1=Dense(10,activation='softmax')(x1)
```

```python
#define the model
model2=Model(inputs1,outputs1)
```

```python
#compile the model
model2.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#early stopping technique
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True,
    verbose=1
)
```

```python
#train the model with Early stopping
history2=model2.fit(
    x_train,y_train,epochs=40,batch_size=64,validation_data=(x_val,y_val),verbose=1,callbacks=[early_stopping]
)
```

```python
#evaluate the model
model2.loss,model2_accuracy=model2.evaluate(x_test,y_test)
print('Accuracy',model2_accuracy)
```

```python
#print accuracy with and without early stopping
plt.figure()
plt.plot(history1.history['accuracy'], label='without_early_stopping')
plt.plot(history2.history['val_accuracy'], label='with_early_stopping')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
