```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn .preprocessing import StandardScaler
from keras.layers import Input,Dense
from keras.models import Model
```

```python
#Load and display the image
img = Image.open('/content/drive/MyDrive/sunflower.jpg')

plt.figure(figsize=(10,6))
plt.imshow(img,cmap='gray')
plt.show()
```

```python
#Convert image to numpy array
image_matrix = np.array(img)
original_dimensions = image_matrix.shape
print("Original dimensions:", original_dimensions)
```

```python
#Flatten the image for PCA
image_matrix_flat = image_matrix.flatten().reshape(-1,1)
```

```python
#Normalise the image data
sc = StandardScaler()
X = sc.fit_transform(image_matrix_flat)
```

```python
#AutoEncoder parameters
encoding_dim = int(0.1 * original_dimensions[0])
```

```python
#AutoEncoder model
input_img = Input(shape=(1,))
encoded = Dense(encoding_dim,activation='linear')(input_img)
decoded = Dense(1,activation='linear')(encoded)
```

```python
# Build the autoencoder
autoencoder = Model(input_img, decoded)

# Compile the model with Mean Squared Error loss
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder
autoencoder.fit(
    X, X,
    epochs=20,
    batch_size=256,
    shuffle=True,
    verbose=1
)

# Reconstruct the image
X_reconstructed = autoencoder.predict(X)

# Calculate Mean Squared Error
mse = np.mean((X - X_reconstructed) ** 2)
print("Mean Squared Error (MSE):", mse)
```

```python
# Inverse normalization
X_reconstructed_inv = sc.inverse_transform(X_reconstructed)

# Reshape back to original image shape
reconstructed_image = X_reconstructed_inv.reshape(original_dimensions)

plt.figure(figsize=(10,6))
plt.imshow(reconstructed_image.astype(np.uint8))
plt.axis('off')
plt.show()
```
